## 문장 토큰화

```python
from nltk.tokenize import sent_tokenize #문장토큰을 쉽게 할 수 있다.
```

```python
# 주어진 문장 단위로 토큰
print(sent_tokenize(sp1))
```

## 단어 토큰화

```python
from nltk.tokenize import word_tokenize
```

```python
# 주어진 단어 단위로 토큰
print(word_tokenize(sp1))
```

- 느낌표나 기타 ' 분리 등 다양하게 분리할 수 있는 패키지

## 정규표현식 토크나이즈

```python
from nltk.tokenize import RegexpTokenizer
```

```python
# 정규표현식으로 제거할 수 있다.
tokenizer = RegexpTokenizer('[\w"]+')
```

```python
print(tokenizer.tokenize(sp1))
#온점, 쉼표, 하이픈 같은 특수문자들이 모두 제거된 뒤, 의미를 가지는 단어들만 남아있는 것을 확인할 수 있음.
```

```python
# stopwords를 제외한 단어만 추출하면 된다.
res=[word for word in tokens if word not in en_stops]
```

## 정규화 작업
- 동사 변화 되는 것
- 흔히 어간(stem) 변하지 않는 부분 이 어간을 추출한다는 개념
- 간다, 갔다 작다, 작으니, 작고 , 작아서 이런 것들 작다의 개념에서 파생되는 것

- 포터 스테머 정규화하는 패키지
- porterStemmer

```python
from nltk.stem import PorterStemmer #정규화 하는 패키지

stem = PorterStemmer()

print(stem.stem('go'), stem.stem('going'), stem.stem('goes'))
```

```python
tokens = word_tokenize(sp1)

#불용어 한 것처럼 반복문으로 진행
res1 = [stem.stem(token) for token in tokens]
```

```python
res1
#단어가 원형으로 바뀌고 불용어가 제거된 것을 확인할 수 있음.
```

```python
tokens
```

```python
lstem = LancasterStemmer()
```

```python
print(lstem.stem('go'), lstem.stem('going'), lstem.stem('goes'))
```

## 표제어 추출 Lemmatization

- 표제어 추출의 경우는 주어진 단어를 기본형을 변환하는 것

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemma = WordNetLemmatizer()
```

```python
print(lemma.lemmatize('go'))
print(lemma.lemmatize('going'))
print(lemma.lemmatize('goes'))


print(lemma.lemmatize('going', pos ='v')) #품사를 지정해서 동사로 뽑을 수 있다.
#기본형과 어간 추출은 그 자체로 차이가 있을듯
```
