```python
import pandas as pd
```

```python
data1 = pd.Series(['1반','2반','3반','4반'])
data2 = pd.Series([1,0,0,4])
```

```python
data1 = data1.append(data2, 'ignore_index = True')
#append -> series와 series를 연결하는 방법, 내부 인자에 다른 series를 넣어주면 세로로 더한다.
```

```python
print(data1)
```

```python
data2.describe()
#describe -> 숫자 data로 이루어진 series에서 여러 통계치들을 반환함
```

```python
data1 = pd.Series(['1반','2반','3반','4반','5반','5반','5반'])
data1 = data1.drop_duplicates()
print(data1)
#drop_duplicates : 중복된 값을 제거함
```

```python
a=data1.values
print(type(a))
print(a)
#values -> series를 numpy로 변환함. 패치된 버전에서는 get_values 대신 values만 허용됨
```

```python
data1 = pd.Series([156,99,1000,4,1])


data1.sort_values()
#sort_values -> 숫자 데이터로 이루어진 series에서 값들을 크기 순으로 배열함
```

```python
data1 = pd.Series([156,99,1000,4,1])


data1.sort_values(ascending = False)
#이런 식으로 내림차순 정렬도 가능하다.
```

```python
data1 = pd.Series(['1반','2반','3반','4반'])
data2 = pd.Series([1,0,0,4])
data1 = data1.append(data2, 'ignore_index = True')
```

```python
data1 = data1.to_frame()
data1
#to_frame -> series를 dataframe으로 변환한다.
```

```python
data1 = pd.Series([156,99,1000,4,1])
print(data1.isin([156]))
#is_in -> 내부에 해당 값이 있는지 확인하여 bool 값을 반환한다
```

```python
def devider(a):
  return a/4


data1.apply(devider)
#apply-> 괄호 안에 함수를 넣어 series 내부 인자에 적용한다. agg와 비슷한 기능이다
```
