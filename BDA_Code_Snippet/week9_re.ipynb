```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
plt.scatter([1,2,3,4],[5,6,7,8]) #범주형 데이터이므로 산점도로 표현하기
plt.title('scatter')
plt.xlabel('x axis')
plt.ylabel('y axis')
```

```python
plt.scatter([1,2,3,4],[5,6,7,8])
plt.xlim(0,4)
plt.ylim(5,8)
plt.title('scatter')
plt.xlabel('x axis')
plt.ylabel('y axis')
#lim을 통해 원하는 범위의 데이터만 표시하기
```

```python
plt.scatter([1,2,3,4],[5,6,7,8])
plt.xlim(right = 3)
plt.ylim(top=4)
plt.title('scatter')
plt.xlabel('x axis')
plt.ylabel('y axis')
#left, top을 통해 역순으로 내려올 수도 있다
#비슷하게 right를 쓰면 오른쪽 끝을 3으로 고정한다.
```

```python
plt.scatter([1,2,3,4],[5,6,7,8])
plt.xticks([1,2,3,4],['A','B','C','D'])
plt.yticks([5,6,7],['data','analy','sis'])

plt.title('scatter')
plt.xlabel('x axis')
plt.ylabel('y axis')
#xticks, yticks -> 데이터 눈금 설정
```

### 범례설정

### plt문법이 겹치게 사용하면 추가 확장된다.

```python
plt.scatter([1,2,3,4],[5,6,7,8], label='cross1')
plt.scatter([1,2,3,4],[8,7,6,5], label='cross2') #층이 쌓여질 수 있다.
plt.title('scatter')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
```

```python
plt.scatter([1,2,3,4],[5,6,7,8], label='x=4')
plt.scatter([1,2,3,4],[8,7,6,5], label='x=3') #층이 쌓여질 수 있다.
plt.title('scatter')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend(loc = 'upper left')
#legend 안쪽에 loc으로 선언해준다.
```

### 색상, 선 스타일, 두께 설정 가능
- color,  linestyle, linewidth 조정할 수 있는 파라미터

```python
plt.figure(figsize=(15,10))
plt.plot([1,2,3],[4,5,6],color='red' ,label= 'x1')
plt.plot([1,2,3],[6,5,4],color='blue',label= 'x2')
plt.legend()
#색상도 마음대로 조정할 수 있다
```

```python
plt.figure(figsize=(15,10))
plt.plot([1,2,3],[4,5,6],color='red',linewidth=8,linestyle='dashed',label= 'x1')
plt.plot([1,2,3],[6,5,4],color='green',linewidth=3,label= 'x2')
plt.legend()
```

```python
plt.figure(figsize=(15,10))
plt.plot([1,2,3],[4,5,6],marker='*',markersize=5,color='black' ,label= 'x1')
plt.plot([1,2,3],[6,5,4],color='blue',label= 'x2')
plt.legend()
#marker: 해당 점을 다양한 모양으로 표시할 수 있음
```

```python
import pandas as pd
df = pd.DataFrame({'+bias':[100,2,1],
                  '-bias':[2,1,5]},
                 index = ['Python','R','SQL'])
#이러면 100을 표시하기 위해 나머지가 모두 뭉개져서 정확한 대소 비교가 안됨
#표준화를 진행하거나 다른 형태의 그래프를 선택
```

```python
sns.heatmap(df)
```

- 데이터 표시하고 싶은 경우
- annot 인자 True
- fmt 사용하면 'd' 정수, 실수 등 표현 가능
- cmap 색상 조합을 다양하게 만들 수 있다.

```python
sns.heatmap(df, annot=True, fmt='.4f') # 정수 소수 값들 다 조정 가능하다!
```

```python
sns.heatmap(df, annot=True, fmt='.1f', linewidth = 2, linecolor = 'orange')
```

```python
sns.heatmap(df, annot=True, fmt='.1f',cmap='Greys', linewidth = 2, linecolor = 'black')
#cmap -> 배경색
```

### 그래프를 실제로 그려보아서 이 축을 지정하여 그래프를 그리자!

```python
x1 = np.arange(1,10)
x2 = np.arange(1,30)
y1 = x1*2
y2 = x2*3
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10,7))

#축을 지정해서 그래프를 그려보자!
ax[0][0].plot(x1,y1, color='red')
ax[1][0].bar(x2,y2, color='green')

ax[0][0].set_xlabel('ax[0] axis')
ax[1][0].set_xlabel('ax[1] axis')

ax[0][0].legend()
ax[1][0].legend()
#이런 식으로 다양하게 분할해서 원하는 그림을 그려볼 수 있다.
```
