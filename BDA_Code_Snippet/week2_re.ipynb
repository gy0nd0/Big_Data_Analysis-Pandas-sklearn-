### 파생변수
- 외부변수, 파생변수
- 통상적으로는 내부 데이터를 가지고 새로운 변수를 만드는 것
- 재료 조건이 내부 데이터 안에 있고 그 데이터를 가지고 -> 새로운 변수를 만든다.


- 그 변수를 왜 만드는가?
- 1. 데이터 분석에 필요한 경우 만들 수 있다. 새로운 평균이라는 기준점을 가지고 분석을 해야 하는 경우 (매출 CAGR 3개년치 성장율)
- 2. 모델링을 통해 피처에 대한 엔지니어링을 진행하는 경우
- 기타 등등..

```python
df['3_mean'] = (df['math']+df['english']+df['science'])/3 #분석에 사용할 수 있도록 기존의 특성을 활용하여 새로운 특성을 만드는
```

```python

df.assign(mean_s = (df['math']+df['english']+df['science'])/3)
```

```python
df.assign(3_mean=(df['math']+df['english']+df['science'])/3)
# 숫자로 시작하는 컬럼명을 쓸 때엔 assign을 쓸 수 없음.
```

### groupby
- 그룹으로 묶는다.
- 그룹으로 통계치를 계산하거나, 카운팅을 계산하거나 등등
- 결국은 그룹으로 무엇을 묶는다는 것이 중요한 포인트!

- df.groupby(기준).agg(변수, 요약통계치) 결과를 확인한다.

```python
df
df.groupby('nclass')
```

```python
df
```

```python
df.groupby('nclass').agg(ncl_math = ('math','max')) #group(그룹화할 대상).agg(이름 = (대상, 방법))
```

```python
df.groupby('nclass').agg(ncl_math = ('math','mean'))#class에 따라 묶고, math의 평균값을 새롭게 반환.
```

```python
df.groupby('nclass').agg(ncl_math = ('math','mean'),
                        ncl_eng = ('english','sum'))#여러 개를 만들 수도 있
```

```python
df_gp=df.groupby('nclass').agg(ncl_math = ('math','std'),
                        ncl_eng = ('english','sum'),
                        ncl_math_mean = ('math','mean'))
```

```python
df
```

```python
df_gp=df.groupby('nclass').agg(ncl_math = ('math','std'),
                        ncl_eng = ('english','sum'),
                        ncl_math_mean = ('math','mean'))
```

```python
df_gp=df.groupby('nclass').agg(ncl_math = ('math','std'),
                        ncl_eng = ('english','max'),
                        ncl_math_mean = ('math','std'))
```
