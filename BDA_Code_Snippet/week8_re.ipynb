## merge 함수
join 개념으로 공통적인 부분을 기준으로 df를 병합한다.


-----
<대표적인 merge 관련 parameter는 다음과 같다.>
- left : 가장 먼저 쓴 데이터프레임
- right : 두 번째 쓴 데이터프레임
- on : join key로 사용할 컬럼 설정
- left_on : 왼쪽 데이터프레임에서 join 키로 사용할 컬럼 목록 적기
- right_on : 오른쪽 데이터프레임에서 join 키로 사용할 컬럼 목록 적기
- left_index : 왼쪽 데이터프레임의 join기준 인덱스 사용할지 여부
- right_index : 오른쪽 데이터프레임 join기준 인덱스 사용할지 여부
- how : join 방법을 결정한다. 기본 디폴트는 'inner'



```python
df1 = pd.DataFrame({'이름':['홍길동','박길동','김길동','이길동'], '반':['파문응','파문기','데분기','데분중']})
df2 = pd.DataFrame({'이름':['홍길동','박길동','오길동','이길동'], '점수':[100,50,80,90]})
#df3 = pd.DataFrame({'A':[7,8,9], 'B':[10,11,12],'C':[13,14,15]})
```

```python
display(df1)
display(df2)
```

```python
pd.merge(df1, df2, on='이름')#이름을 기준으로 두 df를 합친다. 다만, 공통적인 부분만 추출되어 합쳐진다.(inner join이 default이기 때문)
```

```python
pd.merge(df2, df1, on='이름') #순서를 바꾸고 싶다면 입력하는 df 순서를 바꿔주면 된다.
```

- inner 조인 디폴트라-> 공통의 교집합만 출력하게 된다.

### join 키가 두 개 이상인 경우!

```python
df1 = pd.DataFrame({'이름':['홍길동','박길동','김길동','이길동'], '반':['파문응','파문기','데분기','데분중'],'학번':[20,21,22,23],'결석 횟수':[0,0,0,1]})
df2 = pd.DataFrame({'이름':['홍길동','박길동','오길동','이길동'], '점수':[100,50,80,90],'학번':[20,19,22,23],'결석 횟수':[0,1,0,1]})
```

```python
display(df1)
display(df2)
```

```python
pd.merge(df1, df2, on=['이름','학번','결석 횟수'])
```

```python
pd.merge(df1, df2, on=['이름','결석 횟수'])
```

### left_on, right_on


```python
df1 = pd.DataFrame({'이름':['홍길동','박길동','김길동','이길동'], '반':['파문응','파문기','데분기','데분중'],'번호':[20,21,22,23]})
df2 = pd.DataFrame({'이름':['홍길동','박길동','오길동','이길동'], '점수':[100,50,80,90],'학번':[20,19,22,23]})
```

```python
display(df1)
display(df2)
```

```python
#중복이 나오는 경우
pd.merge(df1, df2, left_on='번호', right_on='학번')
#두 df의 컬럼명이 다르지만 data의 내용은 동일한 경우, left_on과 right_on을 이용해서 합쳐줄 수 있다.
```

```python

pd.merge(df1, df2, left_on='이름', right_on='이름')
#컬럼명까지 동일한 경우라면 이렇게 처리하면 되겠지
```

- join key로 잡은 것 외에 이름이라는 두 개의 컬럼이 동일하다.
- 중복이 발생해서 이름_x, 이름_y ( 중복 )

```python
display(df1)
display(df2)
```

```python
#만약 인덱스에 포함된 데이터와 다른 df의 특정 컬럼의 데이터가 동일하다면 어떻게 합칠 수 있을까?

df_sp=df1.set_index('이름')
```

```python
df_sp
```

```python
pd.merge(df_sp, df2, left_index = True, right_on='이름')
#left에선 index의 내용을 기준으로, right에선 이름 컬럼을 기준으로 합친다
```

```python
pd.merge(df_sp, df2, left_index = True) #right_on도 지정해야 한다
```

### join의 다양한 방법 How
<how 인자는 조인 방법을 다양하게 설정할 수 있다.>
- 'inner' - 공통의 교집한 컬럼을 가지고 거기에 교집합된 내용들만 병합하는 것
- 'left' - 왼쪽 데이터프레임의 join 키를 기준으로 병합
- 'right' - 오른쪽 데이터프레임의 Join 키를 기준으로 병합
- 'outer' - 왼쪽과 오른쪽의 데이프레임의 조인키에 대한 합집합을 기준으로 병합

```python
df1 = pd.DataFrame({'이름':['A','B','C','D'], '반':['파문응','파문기','데분기','데분중']})
df2 = pd.DataFrame({'이름':['A','B','E','F'], '점수':[100,50,80,90]})
```

```python
display(df1)
display(df2)
```

```python
pd.merge(df1, df2, on='이름')#inner join이므로 공통된 내용만 추출해서 따로 병합
```

```python
pd.merge(df1, df2, on='이름', how='inner')#위와 동일한 결과
```

```python
pd.merge(df1,df2, on='이름',how='left')
#how를 left로 설정하면, 먼저 적은 df는 일단 고정되고, 거기에 오른쪽 df가 붙는 구조이다.
#만약 해당 데이터가 존재하지 않는다면 na값이 들어간다.
```

```python
pd.merge(df1,df2, on='이름',how='right')
```

```python
display(df1)
display(df2)
```

- outer
- 전체 모든 값을 다 합친 후에 교집합 내용들만 데이터 기입하는 것

```python
pd.merge(df1,df2, on='이름',how='outer') #일단 다 더하고 없는 데이터는 na 처리
```

## 샘플 데이터로 데이터 병합해 보자!

```python
site = pd.read_csv('survey_site.csv')
survey= pd.read_csv('survey_survey.csv')
vit = pd.read_csv('survey_visited.csv')
```

```python
display(site)
display(survey)
display(vit)
```

```python
#컬럼명은 다르지만 동일한 종류의 데이터를 가지는 df인 경우엔 병합할 필요가 있다
#컬럼의 이유가 다르므르 left, right on을 사용한다.
mg1 = pd.merge(site,vit, left_on=['name'], right_on=['site'])
```

```python
print(mg1)
#name과 site를 병합했다
```

```python
mg2=pd.merge(mg1,survey, left_on=['ident'], right_on=['taken'])
print(mg2)
```
