### Groupby

```python
df.groupby('continent').lifeExp.mean()#국가를 기준으로 분류해서 기대수명의 평균값을 추출
```

```python
print(df.groupby('continent')[['lifeExp','pop']].mean())
print(type(df.groupby('continent')[['lifeExp','pop']].mean()))
#2개 이상을 묶으면 데이터프레임으로 출력한다.
```

```python
#groupby + agg -> 임의의 기준을 정해서 컬럼을 분류할 수 있
```

```python
def max(values):
    n = len(values)
    maxer = 0
    for value in values:
        if(value >= maxer):
          maxer = value
        else:
          continue
    return maxer
#최댓값을 구하는 함수를 agg 안에 넣어줄 수 있다.
```

```python
df.groupby('continent').lifeExp.agg(max)
```

```python
import numpy as np
df.groupby('continent').lifeExp.agg([np.mean, np.std, np.count_nonzero])#numpy에서 제공하는 것을 사용할 수도 있다
```

```python
pew= pd.read_csv('pew.csv')
```

```python
pew #melt 함수를 이용해서 깔끔하게 정리해보자
```

```python
pd.melt(pew,id_vars='religion') #id_vars에 넣는 값은 그대로 유지 -> 나머지 컬럼을 하나씩 떼어서 아래로 늘어놓음
```

```python
pd.melt(pew,id_vars='religion',var_name='income',value_name='count')
#각 컬럼의 명칭을 새롭게 지정하는 parameter: var_name
#각 컬럼의 값을 집어넣는 새로운 변수 parameter: value_name
```

```python
np.transpose(pew) #단순 축만 바꾸는 개념
```
